```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import openpyxl
# Create a DataFrame

all_song_titles = []

start_url = 'https://akb48.fandom.com/wiki/Category:Songs'

base_url = 'https://akb48.fandom.com'
#start_url = 'https://akb48.fandom.com/wiki/Category:NMB48_Songs'

def get_soup(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # This will raise an HTTPError if the HTTP request returned an unsuccessful status code
        return BeautifulSoup(response.content, 'html.parser')
    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return None

def scrape_songs(soup):
    song_items = soup.select('.category-page__member .category-page__member-link')
    return [item.get_text(strip=True) for item in song_items]

def find_next_page(soup):
    next_button = soup.select_one('.category-page__pagination-next')
    return next_button['href'] if next_button else None

current_url = start_url

while current_url:
    print(f"Scraping page: {current_url}")
    soup = get_soup(current_url)
    if soup:
        songs = scrape_songs(soup)
        current_url = find_next_page(soup)
        all_song_titles.extend(songs)
    else:
        break
df = pd.DataFrame({'Song Title': all_song_titles})
print(df)
df.index += 1
df.index.name = 'No.'
    
#display(df)
# Export to CSV
df.to_excel('48Gsongs_all.xlsx', index=True)
#print("Exported to 48Group songs.csv successfully.")
```
